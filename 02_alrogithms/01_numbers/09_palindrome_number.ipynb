Given an integer x, return true if x is a palindrome, and false otherwise.

Example 1:
- Input: x = 121
- Output: true
Explanation: 121 reads as 121 from left to right and from right to left.

Example 2:
- Input: x = -121
- Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.

Example 3:
- Input: x = 10
- Output: false
Explanation: Reads 01 from right to left. Therefore it is not a palindrome.
 

Constraints:
-231 <= x <= 231 - 1
 
Follow up: Could you solve it without converting the integer to a string?

```python
def is_palindrome(x):
    """
    :type x: int
    :rtype: bool
    """
    return str(x) == str(x)[::-1]
```

```python
import unittest


class TestStringSollution(unittest.TestCase):
    def test_plaindrome_number_1(self):
        self.assertEqual(is_palindrome(121), True)

    def test_plaindrome_number_2(self):
        self.assertEqual(is_palindrome(-121), False)

    def test_plaindrome_number_3(self):
        self.assertEqual(is_palindrome(10), False)


unittest.main(argv=[""], verbosity=2, exit=False)
```

```python
def is_palindrome_no_string(x: int) -> bool:
    """
    :type x: int
    :rtype: bool
    """
    if x < 0:
        # negative numbers cannot be palindromes
        return False

    reverse = 0  # init var to store reverse of input number
    xcopy = x  # copy of original input nmber

    # wtart a while loop that continues until the input number x becomes 0.
    # in each iteration:
    #  - get the last digit of x by taking the remainder when divided by 10 (this is x % 10).
    #  - add this last digit to the reverse variable after shifting the existing digits one place to the left (reverse * 10).
    #  - update x by removing its last digit (integer division by 10: x //= 10).
    while x > 0:
        reverse = (reverse * 10) + (x % 10)
        x //= 10

    return reverse == xcopy
```

```python
import unittest


class TestNoStringSollution(unittest.TestCase):
    def test_plaindrome_number_1(self):
        self.assertEqual(is_palindrome_no_string(121), True)

    def test_plaindrome_number_2(self):
        self.assertEqual(is_palindrome_no_string(-121), False)

    def test_plaindrome_number_3(self):
        self.assertEqual(is_palindrome_no_string(10), False)


unittest.main(argv=[""], verbosity=2, exit=False)
```
