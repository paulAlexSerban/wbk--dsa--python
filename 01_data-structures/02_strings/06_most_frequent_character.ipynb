## Reference Solution

```python
def most_frequent_char(s: str) -> str:
    if not s:
        raise ValueError("Input string must not be empty")

    # Step 1: build frequency map
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    # Step 2: find the max-frequency character.
    # Iterating over s (not freq) ensures that among ties the first-occurring
    # character is naturally selected — we only update when strictly greater.
    best_char = s[0]
    for ch in s:
        if freq[ch] > freq[best_char]:
            best_char = ch

    return best_char
```
