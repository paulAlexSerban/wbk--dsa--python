## Reference Solution

```python
def word_frequencies(text: str) -> dict:
    freq = {}
    # str.split() with no args splits on any run of whitespace,
    # so "   " produces [] — the loop simply never executes.
    for word in text.lower().split():
        # dict.get(key, 0) returns 0 when the word is seen for the first time,
        # avoiding a KeyError without needing an explicit if/else.
        freq[word] = freq.get(word, 0) + 1
    return freq
```
