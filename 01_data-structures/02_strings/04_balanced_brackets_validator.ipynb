## Real-World Context

Balanced-bracket validation is at the core of every compiler, linter, and code editor. When your IDE highlights a mismatched brace or your JSON parser throws a syntax error, the underlying check is almost identical to what you are implementing here. The stack data structure — "last in, first out" — is the natural fit because the most recently opened bracket must be the next one closed.

## Reference Solution

```python
def is_balanced(s: str) -> bool:
    # Maps each closing bracket to the opening bracket it must match.
    matching = {')': '(', ']': '[', '}': '{'}

    stack = []  # acts as a LIFO stack; Python lists support O(1) append/pop from the end.

    for ch in s:
        if ch in "([{":
            # Opening bracket: push onto stack and wait for its closing pair.
            stack.append(ch)
        elif ch in ")]}":
            # Closing bracket: the top of the stack must be its matching opener.
            # An empty stack means there is no opener to match — immediately unbalanced.
            if not stack or stack[-1] != matching[ch]:
                return False
            stack.pop()  # consume the matched opener
        # Any other character (letters, digits, spaces) is silently ignored.

    # If the stack is empty, every opener was matched; otherwise some are still open.
    return len(stack) == 0
```
