## Expected Output / Test Cases

```python
# Helper: order-independent comparison of grouped results
def normalise(groups):
    return sorted([sorted(g) for g in groups])

# Happy path
result = group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"])
assert normalise(result) == normalise([["eat", "tea", "ate"], ["tan", "nat"], ["bat"]])

# Single word — forms its own group
assert normalise(group_anagrams(["hello"])) == [["hello"]]

# All words are anagrams of each other
assert normalise(group_anagrams(["abc", "bca", "cab"])) == [["abc", "bca", "cab"]]

# No words share letters — each forms its own group
result = group_anagrams(["dog", "cat", "log"])
assert normalise(result) == normalise([["dog"], ["cat"], ["log"]])

# Edge case — empty input
assert group_anagrams([]) == []

# Edge case — list with one word
assert group_anagrams(["x"]) == [["x"]]

# Case-sensitive: "Eat" ≠ "eat"
result = group_anagrams(["Eat", "eat"])
assert normalise(result) == normalise([["Eat"], ["eat"]])
```

## Real-World Context

Grouping by a computed signature is a pattern that appears constantly in real systems: deduplicating database records, clustering similar documents, routing identical jobs to cached results, and even finding collision groups in hash tables. ## Reference Solution

```python
from collections import defaultdict

def group_anagrams(words: list) -> list:
    # defaultdict(list) automatically creates an empty list for any new key,
    # so we never need to check "does this key exist yet?".
    groups = defaultdict(list)

    for word in words:
        # Sorting the characters produces a canonical key:
        # "eat" → ('a','e','t'), "tea" → ('a','e','t'), "ate" → ('a','e','t')
        # A tuple is used (not a list) because dictionary keys must be hashable.
        key = tuple(sorted(word))
        groups[key].append(word)

    # dict.values() returns the groups; wrapping in list() gives the required type.
    return list(groups.values())
```
