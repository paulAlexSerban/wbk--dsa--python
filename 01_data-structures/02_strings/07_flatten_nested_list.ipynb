## Real-World Context

One-level flattening is a common data-wrangling step: API responses often return paginated batches (a list of lists of records), configuration parsers may yield grouped option blocks, and event-log processors receive chunked sequences. Being explicit about *one level only* matters because over-flattening structured data can destroy hierarchical information you still need downstream.

## Reference Solution

```python
def flatten(nested: list) -> list:
    result = []
    for item in nested:
        if isinstance(item, list):
            # extend adds each element of the inner list individually —
            # this is exactly one level of flattening; inner structure of
            # item is preserved (e.g. [3, 4] inside item stays [3, 4]).
            result.extend(item)
        else:
            # Non-list items (integers, strings, etc.) are kept as-is.
            result.append(item)
    return result
```
