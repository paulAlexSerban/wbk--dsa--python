## Expected Output / Test Cases

```python
# Happy path — encode
assert caesar("hello", 3) == "khoor"
assert caesar("xyz", 3) == "abc"           # wrap-around
assert caesar("Hello, World!", 13) == "Uryyb, Jbeyq!"  # mixed case + punctuation

# Decode
assert caesar("khoor", 3, decode=True) == "hello"
assert caesar("Uryyb, Jbeyq!", 13, decode=True) == "Hello, World!"

# ROT-13 is its own inverse
encoded = caesar("Python", 13)
assert caesar(encoded, 13, decode=True) == "Python"

# Shift of 0 — no change
assert caesar("abc", 0) == "abc"

# Shift larger than 26 — normalised
assert caesar("a", 27) == "b"   # 27 % 26 == 1

# Negative shift
assert caesar("d", -3) == "a"

# Edge cases
assert caesar("", 5) == ""
assert caesar("123 !?", 10) == "123 !?"   # no alphabetic chars → unchanged
```

## Real-World Context

The Caesar cipher is the ancestor of modern symmetric encryption. While trivially broken today, it introduces the two ideas at the heart of every cipher: a **key** (the shift value) and a **reversible transformation** (encode/decode with the same key). The `ord`/`chr` arithmetic you practise here — converting characters to integers, doing modular arithmetic, and converting back — appears in Unicode processing, checksum calculation, and protocol serialisation.

## Reference Solution

```python
def caesar(text: str, shift: int, decode: bool = False) -> str:
    # When decoding, negate the shift; normalise to 0–25 with % 26 so
    # negative and large-positive shifts both work correctly.
    effective_shift = (-shift if decode else shift) % 26

    result = []
    for ch in text:
        if ch.islower():
            # Map to 0–25, shift, wrap, map back to lowercase ASCII.
            shifted = (ord(ch) - ord('a') + effective_shift) % 26
            result.append(chr(shifted + ord('a')))
        elif ch.isupper():
            # Same logic, but anchored to ord('A').
            shifted = (ord(ch) - ord('A') + effective_shift) % 26
            result.append(chr(shifted + ord('A')))
        else:
            # Non-alphabetic characters pass through unchanged.
            result.append(ch)

    # join is O(n) and avoids repeated string concatenation.
    return "".join(result)
```
