## Problem Statement

Write a function `caesar(text: str, shift: int, decode: bool = False) -> str` that applies (or reverses) a Caesar cipher to `text`.

A **Caesar cipher** shifts each letter in the alphabet by a fixed number of positions. With `shift=3`, `'a'` becomes `'d'`, `'z'` wraps around to `'c'`, and so on.

**Requirements:**
- Shift only **alphabetic** characters (`A–Z`, `a–z`); leave all other characters (spaces, digits, punctuation) unchanged.
- Preserve **case**: uppercase letters shift within `A–Z`, lowercase within `a–z`.
- Handle **wrap-around**: shifting `'z'` by 3 gives `'c'`, not a character outside the alphabet.
- When `decode=True`, reverse the shift to recover the original text.
- `shift` may be any integer (positive, negative, or zero); normalise it with `% 26` so out-of-range values work correctly.
- An empty string returns an empty string.


````python
def caesar(text, shift, decode = False):
    if not isinstance(shift, int):
        raise ValueError("Shift must be an integer.")
    if not isinstance(text, str):
        raise ValueError("Text must be a string.")
    if shift < 1 or shift > 25:
        raise ValueError("Shift must be between 1 and 25.")
    
    alphabet = 'abcdefghijklmnopqrstuvwxyz'

    if decode:
        # If decoding, we need to shift in the opposite direction
        # this decodes by shifting in the opposite direction, which is equivalent to encoding with a negative shift
        shift = -shift

    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    translation_table = str.maketrans(alphabet + alphabet.upper(), shifted_alphabet + shifted_alphabet.upper())
    return text.translate(translation_table)
````

## Expected Output / Test Cases

````python
import unittest


class TestStringSollution(unittest.TestCase):
    def test_caesar_happy_path_encode(self):
        # Happy path — encode
        self.assertEqual(caesar("hello", 3), "khoor")
        self.assertEqual(caesar("xyz", 3), "abc")
        self.assertEqual(caesar("Hello, World!", 13), "Uryyb, Jbeyq!")

    def test_caesar_happy_path_decode(self):
        # Decode
        self.assertEqual(caesar("khoor", 3, decode=True), "hello")
        self.assertEqual(caesar("Uryyb, Jbeyq!", 13, decode=True), "Hello, World!")

    def test_caesar_rot13(self):
        # ROT-13 is its own inverse
        encoded = caesar("Python", 13)
        self.assertEqual(caesar(encoded, 13, decode=True), "Python")

unittest.main(argv=[""], verbosity=2, exit=False)
````

## Real-World Context

The Caesar cipher is the ancestor of modern symmetric encryption. While trivially broken today, it introduces the two ideas at the heart of every cipher: a **key** (the shift value) and a **reversible transformation** (encode/decode with the same key). The `ord`/`chr` arithmetic you practise here — converting characters to integers, doing modular arithmetic, and converting back — appears in Unicode processing, checksum calculation, and protocol serialisation.

## Reference Solution

````python
def caesar(text: str, shift: int, decode: bool = False) -> str:
    if decode:
        shift = -shift
    shift = shift % 26  # Normalise shift to be within [0, 25]

    result = []
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            shifted = (ord(char) - base + shift) % 26
            result.append(chr(shifted + base))
        else:
            result.append(char)
    return ''.join(result)
````

````python
import unittest


class TestStringSollution(unittest.TestCase):
    def test_caesar_happy_path_encode(self):
        # Happy path — encode
        self.assertEqual(caesar("hello", 3), "khoor")
        self.assertEqual(caesar("xyz", 3), "abc")
        self.assertEqual(caesar("Hello, World!", 13), "Uryyb, Jbeyq!")

    def test_caesar_happy_path_decode(self):
        # Decode
        self.assertEqual(caesar("khoor", 3, decode=True), "hello")
        self.assertEqual(caesar("Uryyb, Jbeyq!", 13, decode=True), "Hello, World!")

    def test_caesar_rot13(self):
        # ROT-13 is its own inverse
        encoded = caesar("Python", 13)
        self.assertEqual(caesar(encoded, 13, decode=True), "Python")

    def test_caesar_shift_of_0(self):
        # Shift of 0 — no change
        self.assertEqual(caesar("abc", 0), "abc")

    def test_caesar_shift_larger_than_26(self):
        # Shift larger than 26 — normalised
        self.assertEqual(caesar("a", 27), "b")   # 27 % 26 == 1

    def test_caesar_negative_shift(self):
        # Negative shift
        self.assertEqual(caesar("d", -3), "a")

    # Edge cases
    def test_caesar_empty_string(self):
        self.assertEqual(caesar("", 5), "")

    def test_caesar_non_alphabetic_chars(self):
        self.assertEqual(caesar("123 !?", 10), "123 !?")   # no alphabetic chars → unchanged

unittest.main(argv=[""], verbosity=2, exit=False)
````

````python
def caesar(text: str, shift: int, decode: bool = False) -> str:
    # When decoding, negate the shift; normalise to 0–25 with % 26 so
    # negative and large-positive shifts both work correctly.
    effective_shift = (-shift if decode else shift) % 26

    result = []
    for ch in text:
        if ch.islower():
            # Map to 0–25, shift, wrap, map back to lowercase ASCII.
            shifted = (ord(ch) - ord('a') + effective_shift) % 26
            result.append(chr(shifted + ord('a')))
        elif ch.isupper():
            # Same logic, but anchored to ord('A').
            shifted = (ord(ch) - ord('A') + effective_shift) % 26
            result.append(chr(shifted + ord('A')))
        else:
            # Non-alphabetic characters pass through unchanged.
            result.append(ch)

    # join is O(n) and avoids repeated string concatenation.
    return "".join(result)
````
