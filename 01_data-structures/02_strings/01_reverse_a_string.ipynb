## Problem Statement

Write a function `reverse_string(s: str) -> str` that returns the characters of the input string in reverse order.

**Requirements:**
- If `s` is an empty string `""`, return `""`.
- If `s` contains only one character, return that character unchanged.
- Do **not** use any built-in `reverse()` function or `[::-1]` slice shortcut — implement the reversal with an explicit loop so you practise iterating over strings.

````python
def reverse_string(s: str) -> str:
    return s[::-1]
````

## Expected Output / Test Cases

## Reference Solution

```python
def reverse_string(s: str) -> str:
    result = ""
    # Iterate from the last index down to 0 (inclusive).
    # range(start, stop, step) — stop is exclusive, so -1 means "include index 0".
    for i in range(len(s) - 1, -1, -1):
        result += s[i]
    return result
```
